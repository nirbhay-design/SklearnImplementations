Calculate the probability for each word in a given class.

Class 0

```python
prob_eachword_class0 = dict(zip(list(f_0.keys()),list(np.array(list(f_0.values()))/total_f0)))
```

Class 1

```python
prob_eachword_class1 = dict(zip(list(f_1.keys()),list(np.array(list(f_1.values()))/total_f1)))
```

We have calculated the probability of occurrence of word in a class, we can now substitute the values in the Baye's equation. 

If a word from the new sentence does not occur in the class within the training set, the equation becomes zero. This problem can be solved using smoothing like Laplace smoothing. 

```python
#laplace smoothing is done while calculating the probabilities
```

Probability for class 0

```python
p_0 = df_0.shape[0]/(df_0.shape[0]+df_1.shape[0])
p_0
```

Probability for class 1

```python
p_1 = 1-p_0
p_1
```

Print target class

```python

l = []
prob_0 = []
prob_1 = []

for i in test_df.filtered_text:
  a1 = i.split(' ')
  posterior0 = p_0
  posterior1 = p_1
  for j in a1:
    if f_0.get(j,-1) == -1:
      posterior0 *= (1/(total_f0+total_uniquewords)) #laplace smoothing
    else:
      posterior0 *= prob_eachword_class0.get(j,0)#(f_0.get(j,0)/total_f0)
    if f_1.get(j,-1) == -1:
      posterior1 *= (1/(total_f1+total_uniquewords)) #laplace smoothing
    else:
      posterior1 *= prob_eachword_class1.get(j,0)#(f_1.get(j,0)/total_f1)
  prob_0.append(posterior0)
  prob_1.append(posterior1)  
  if (posterior0>posterior1):
    l.append(0)
  else:
    l.append(1)
test_df['predicted'] = l
```
